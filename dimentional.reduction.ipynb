### Dimensionality Reduction using t-SNE

```python
from keras.datasets import mnist
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
(X, y), (_,_) = mnist.load_data()
print(X.shape, y.shape)
```

```python
X = X.reshape(len(X), -1)
X.shape
```

Basically, we are considering the data-set of mnist which is multi-dimension data set with a shape of (60000,784) so our intension is to reduce the dimension to understandable form like 2D or 3D 

# t-SNE

```python
x_tsne = TSNE().fit_transform(X[:10000])
```

Take only 10k samples for quick results and you can run the data with respect to your preference of the CPU and GPU cores and you can go for the CUDA toolkit for quick responses

```python
plt.figure(figsize=(10,10))
sc = plt.scatter(x_tsne[:, 0], x_tsne[:, 1], c=y[:10000])
plt.legend(handles=sc.legend_elements()[0], labels=list(range(10)))
plt.show()
```

Here in the output plot the elements are grouped as per their category  

### let's consider PCA model for understanding variation 

```python
from sklearn.decomposition import PCA
x_pca = PCA(n_components=2).fit_transform(X)
plt.figure(figsize=(10,10))
sc = plt.scatter(x_pca[:, 0], x_pca[:, 1], c=y)
plt.legend(handles=sc.legend_elements()[0], labels=list(range(10)))
plt.show()
```
